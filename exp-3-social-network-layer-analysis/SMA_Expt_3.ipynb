```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(['a', 'b', 'c', 'd'])
```

```python
G.nodes()
```

```python
G.add_edge('a', 'b')
G.add_edge('a', 'c')
G.add_edge('b', 'd')
```

```python
G.edges()
```

```python
nx.draw(G)
plt.show()
```

```python
nx.degree(G)
```

```python
nx.degree_centrality(G)
```

```python
nx.shortest_path(G, 'a', 'd')
```

```python
nx.betweenness_centrality(G)
```

```python
nx.eigenvector_centrality(G)
```

## Random Graph

```python
G = nx.fast_gnp_random_graph(10, 0.5)
nx.draw(G)
```

```python
nx.draw(G, with_labels=1)
```

```python
nx.degree(G)
```

```python
nx.degree_centrality(G)
```

Identifying Most Influential Node

```python
m_influential = nx.degree_centrality(G)
```

```python
for w in sorted(m_influential, key=m_influential.get, reverse=True):
  print(w, m_influential[w])
```

## Activity

```python
G = nx.Graph()
```

```python
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])
```

```python
G.nodes()
```

```python
G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('B', 'D')
G.add_edge('C', 'E')
G.add_edge('D', 'E')
G.add_edge('D', 'G')
G.add_edge('E', 'H')
G.add_edge('E', 'G')
G.add_edge('E', 'F')
G.add_edge('G', 'H')
```

```python
G.edges()
```

```python
nx.draw(G, with_labels='A')
plt.show()
```

```python
nx.degree(G)
```

```python
nx.degree_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

```python
nx.eigenvector_centrality(G)
```

```python
m_influential = nx.degree_centrality(G)
```

```python
for w in sorted(m_influential, key=m_influential.get, reverse=True):
  print(w, m_influential[w])
```

```python
n_influential = nx.betweenness_centrality(G)
```

```python
for w in sorted(n_influential, key=n_influential.get, reverse=True):
  print(w, n_influential[w])
```
